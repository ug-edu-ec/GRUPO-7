```python
import os
```

```python
def convert(input, output):
  exporter = nbconvert.PythonExporter()
  body, resources = exporter.from_filename(path_modulos_ipynb+input)
  with open(path_modulos_py+output, 'w') as f:
      f.write(body)
  f.close()
```

```python
for name in os.listdir(path_modulos_ipynb):
  index = name.index('.')
  name = name[:index]
  convert(name+'.ipynb',name+'.py');
sys.path.append(path_modulos_py)
```
